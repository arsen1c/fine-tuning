```python
# Load IMDb dataset from hugginface datasets library
from datasets import load_dataset

imdb = load_dataset("imdb")
```

```python
"""
There are two fields in this dataset:

    text: the movie review text.
    label: a value that is either 0 for a negative review or 1 for a positive review.

"""
```

```python
# Preprocess

# Load DistilBERT tokenizer to preprocess text field
from transformers import AutoTokenizer

model = "distilbert/distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
tokenizer # prints the whole DistilBERT model
```

```python
tokenizer("Hello world!") # tokenizer output example
```

```python
# preprocessing function to tokenize "text"
def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True)
```

```python
# use Datasets map to apply preprocessing function over the entire dataset
tokenized_imdb = imdb.map(preprocess_function, batched=True) #batch the whole process for speed
```

```python
# Dynamically pads sequences in each batch to the length of the longest sequence,
# ensuring tensors in the batch have the same size before being fed to the model.

from transformers import DataCollatorWithPadding
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# Evaluate
# Including metric during training to help evaluating model's performance
import evaluate

accuracy = evaluate.load("accuracy")
```

```python
# function to pass predictions and labels to calculate accuracy
import numpy as np


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
# before  you train the model, create a map of the expected ids
# to their labels with id2label and label2id
id2label = {0: "NEGATIVE", 1: "POSITIVE"}
label2id = {"NEGATIVE": 0, "POSITIVE": 1}
```

```python
# training your model

# load DistilBERT with AutoModelForSequenceClassification
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert/distilbert-base-uncased", num_labels=2, id2label=id2label, label2id=label2id
)
```

```python
# Define your training hyperparameters in TrainingArguments. T
training_args = TrainingArguments(
    output_dir="../models/imdb_sentiment",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    fp16=True, # make training faster by switch to 16-bit float
    dataloader_num_workers=4 #make training faster
    # push_to_hub=True, push to hugging face
)

# Pass the training arguments to Trainer along with the model, dataset, tokenizer, data collator, and compute_metrics function.
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_imdb["train"],
    eval_dataset=tokenized_imdb["test"],
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
# run the trainer
trainer.train()
```
